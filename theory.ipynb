# Основы написания функций

В этом уроке рассмотрим **функции** — структуры языка Python, которые позволяют уменьшать объем программного кода за счет оформления повторяющихся фрагментов в виде лаконичных названий.

Также рассмотрим возможность изменения работы функции без изменения ее тела — **параметризацию**.

### Ключевые слова, понятия и выражения






*   функция;
*   глобальная функция;
*   локальная функция;
*   лямбда-функция;
*   функция-генератор;
*   передача параметров функции;
*   область видимости переменных.

### Виды функций в Python

В языке Python можно создать четыре типа функций:

1.   **Глобальные функции** — типичный представитель семейства функций. Виден из любой точки программного кода в том же модуле, в котором находится сама функция. На первых порах мы чаще всего будем работать именно с этим типом функций.
2.   **Локальные функции** (их еще называют вложенными функциями) —
это функции, которые объявляются внутри других функций.
3.   **Лямбда-функции** — функции, которые могут создаваться непосредственно в месте их использования. Рассмотрим их в разделе «Лямбда-функции и функции-генераторы».
4.   **Методы** — функции, ассоциированные с определенным типом данных. Мы подробно рассмотрим их в материале по объектно-ориентированному программированию.


 

### Общий вид глобальной функции

Глобальные функции повсеместно используются, но их вид со стороны практически всегда одинаков:
```
def functionName(parameters):
  suite
```


*   ```functionName``` — название или имя функции, по которому мы будем вызывать и использовать функцию в дальнейшем;
*   ```parameters``` — параметры  необязательны и при наличии 
более одного параметра записываются как последовательность 
идентификаторов через запятую или в виде последовательности пар ```identifier=value```;

*   ```suite``` — тело функции, то есть последовательность выполнения определенных действий. Код, выполняемый внутри функции.


 


Чтобы создать функцию:

1. Указываем `def` — ключевое слово, указывающее на то, что дальше пойдет функция. 
2. После `def` ставим пробел и прописываем имя функции: в нашем примере это `functionName`. 
3. Сразу после имени функции ставим круглые скобки `()`, внутри которых указываем параметры функции — те переменные, которые мы подадим внутрь функции. Если для работы функции не требуется передача параметров извне и всё, что нужно, есть в самой функции — скобки оставляем пустыми. 
4. Сразу после `()` ставим двоеточие `:` Всё, что последует дальше, будет телом функции — ее основным кодом. Обратите внимание на отступы: всё тело функции имеет отступ в два пробела относительно строки с `def`. Если мы не будем соблюдать этот отступ, весь код будет распознан как обычный, не принадлежащий функции, и получится некорректный результат.

````python
#Пример глобальной функции

def summ():  # Определяем имя функции
  a = 1 + 2  # Код тела функции
  print(a)   # Отображаем результат

summ()  # Вызываем функцию к исполнению
````

Очень важно то, что прописанное под `def` — это что-то вроде инструкции того, как функция должна работать. Запуск этого кода не приведет к срабатыванию функции, это только определит существование функции, и что она делает при вызове. Чтобы запустить функцию, вызовите ее, обратившись через её имя со скобками: `summ()`. Когда исполнение кода дойдет до `summ()`, функция запустится и код, указанный под `def`, будет исполнен.

### Общий вид локальной функции

Локальные функции помогают реализовывать сложные задачи в глобальной функции, взяв на себя маленькую подзадачу. Важно то, что чем меньше функция будет брать на себя задач, тем легче ее написать и протестировать. Если при этом разместить ее в области видимости глобальной функции, где она только и применяется, можно получить удобный скрытый инструмент.

```
def GlobalfunctionName(global_parameters):
  ...
  def LocalfunctionName(local_parameters):
    suite
  ...
```

Создание локальной функции аналогично созданию глобальной, за тем лишь исключением, что мы создаём ее с отступами внутри тела глобальной функции. Тело локальной функции также указывается с отступом относительно строки, определяющей создание локальной функции. Несоблюдение отступов приведет к тому, что код тела локальной функции будет воспринят как код тела глобальной функции, и некорректно интерпретируется.

### Что и как подавать на вход функции

Для изменения поведения функции необходимо использовать параметры, которые подаются ей на вход. Посмотрим, как эти параметры можно передавать в функцию.

**Позиционные аргументы**

Для них важна последовательность значений, подаваемых в функцию. В какой последовательности значения подаем, в такой последовательности параметры функции примут эти значения.

````python
# Подсчет периметра треугольника

def perimetr(a, b, c):  # Создаем функцию подсчета периметра треугольника perimetr, которая принимает три параметра: a, b и c 
  p = a + b + c         # Выполняем действие в теле функции
  print(p)              # Отображаем результат
  
perimetr(3,4,5)         # Вызываем функцию perimetr и подаем в нее три параметра со значениями 3, 4 и 5
````

**Все позиционные сразу — распаковка аргументов**

Если мы не знаем, сколько параметров должна принять функция, мы можем указать некоторый набор параметров через `*`: 

````python
# Подсчет суммы чисел

def my_sum(*elements):      # Создаем функцию подсчета чисел, которая примет заранее неопределенное число параметров
  res = 0                   # Переменная res будет содержать результат суммирования элементов
  for i in elements:        # В цикле проходим по всем параметрам, поданным в функцию
    res += i                # Добавляем значение параметра в переменную res
  print(elements,'->',res)  # Выводим результаты

my_sum(1)                   # Делаем вызовы функции, подавая в нее разное количество чисел
my_sum(1,3)
my_sum(1,4,5,6)
my_sum(1,3,7,9,12)
````

**Именованные аргументы и аргументы по умолчанию**

Поскольку использовать позиционные аргументы не всегда приемлемо, мы используем именованные аргументы. Значение будет присвоено аргументу, чье имя указано, а не первому по очередности. Мы также можем создать значения по умолчанию, которые будут использованы, если в функцию не будут поданы соответствующие значения.

Данная опция удобна, т. к. не требуется ввод всех параметров для работы функции, а ручная настройка доступна для более опытных пользователей.

````python
# Площадь прямоугольника

def sqr_pr(a=2,b=4):    # Создаем функцию sqr_pr с параметрами, заданными по умолчанию
  s = a * b             # Находим площадь прямоугольника s
  print(s)              # Отображаем результат
  
sqr_pr()                # Вызов функции с параметрами по умолчанию
sqr_pr(a=3)             # Вызов функции: а присваиваем значение 3, b=4 по умолчанию
sqr_pr(a=4,b=8)         # а присваиваем значение 4, b присваиваем значение 8
sqr_pr(4,8)             # первому аргументу присваиваем значение 4, второму присваиваем значение 8
sqr_pr(4)               # первому аргументу присваиваем значение 4, b=4 по умолчанию
````

**Все именованные сразу — распаковка отображений**

Можно создать словарь из именованных параметров и их значений и подать в функцию. В таком случае в качестве параметров функции указываем `**args`:

````python
# Вывод всех параметров функции

def param_list(**args):       # Функция param_list примет словарь именованных аргументов
  print(args)                 # Отобразим то, какие параметры получила функция

param_list(a=2, b=4, c=6)     # Вызов функции с именованными параметрами
````

### Что ждать на выходе функции

До сих пор мы писали функции, которые ничего не возвращали, а только выводили что-то через `print()`. Но в большинстве случаев они нужны для каких-либо подсчетов, и нам необходимо, чтобы функция вернула результат своих вычислений. Для этого в конце функции используют ключевое слово `return`.

**Возращение значений — ```return```**

````python
# Площадь прямоугольника

def sqr_pr(a=2,b=4):      # Функция sqr_pr с параметрами длины и ширины по умолчанию
  s = a * b               # Считаем площадь прямоугольника
  return s                # return возвращает значение площади прямоугольника

sq = sqr_pr(3,b=5)        # Создаем переменную sq, в которую будет помещен результат работы функции
print(sq)                 # Смотрим результат работы функции
````

Через `return` мы можем сразу получить из функции набор нужных нам значений переменных, достаточно указать их через запятую.

````python
# Площадь и периметр прямоугольника

def sqr_pr(a=2,b=4):      # Функция sqr_pr с параметрами длины и ширины по умолчанию
  s = a * b               # Считаем площадь прямоугольника
  p = 2*(a + b)           # Считаем периметр прямоугольника
  return s,p              # return возвращает значение площади и периметра прямоугольника

s, p = sqr_pr(4,6)        # Создаем переменные s и p, в которые будет помещен результат работы функции — площадь и периметр
print('s=',s)             # Смотрим площадь
print('p=',p)             # Смотрим периметр
````

### Документирование функции


Хороший тон, а чаще даже обязательное правило, это добавление описания к любой функции, используя строки документирования. Это обычные строки, которые следуют сразу за строкой с инструкцией `def` и перед программным кодом функции.

Документирование необходимо для того, чтобы другой специалист смог быстро и легко понять ваш код. Или даже вы сами после большого промежутка времени.

Описание работы функции принято открывать тремя одинарными кавычками `'''` и так же закрывать. Внутри описываем, что функция делает, что принимает на вход и что возвращает на выходе. Можно также указать тип данных входных и выходных параметров.

````python
# Площадь и периметр прямоугольника

def sqr_pr(a=2,b=4):     # Функция sqr_pr с параметрами длины и ширины по умолчанию
  '''
     sqr_pr — подсчет площади и периметра прямоугольника
     вход:
         a — длина
         b — ширина
     выход:
         (s,p) — площадь и периметр прямоугольника
  '''
  s = a * b             # Считаем площадь прямоугольника
  p = 2*(a + b)         # Считаем периметр прямоугольника
  return s,p            # return возвращает значение площади и периметра прямоугольника
````

````python
help(sqr_pr)            # через команду help можно вывести документацию к функции
````

# Лямбда-функции и функции-генераторы

Рассмотрим частные случаи функций в Python — лямбда-функции (лямбда-выражения) и функции-генераторы, которые позволяют сделать код кратким и эффективным.

### Ключевые слова, понятия и выражения






*   функция;
*   лямбда-функция;
*   функция-генератор.

### Что такое лямбда-функции и зачем они нужны

**Лямбда-функция** — это функция, которая используется только в одном месте и выполняет одну-единственную задачу.

Лямбда-функции используют для реализации короткого кода функции без необходимости создания для нее длинного описания и переменной, как для обычных функций. Часто их используют в виде подфункций в функциях более высокого порядка, то есть в виде локальных функций.

**Плюсы**: лаконичность (пишем в одну строку, `def` не нужен), автоматический возврат значения.

**Минусы**: сложный для новичка синтаксис, ухудшение читаемости кода, ограниченная функциональность по сравнению с обычными функциями.

### Общий вид лямбда-функций

Для создания лямбда-функций используют следующий синтаксис:

```lambda parameters: expression```, где:


*   ```lambda``` — ключевое слово;
*   ```parameters``` — параметры необязательны, но если присутствуют, то обычно представляют собой простой список имен переменных, разделенных запятыми, то
есть позиционных аргументов. Хотя при необходимости допускается использовать полный синтаксис определения аргументов, используемый в инструкции `def` для глобальных переменных;
*   ```expression``` — тело функции (выражение). Лямбда-функция возвращает результат данного выражения.



### Использование лямбда-функций

**Простой пример**

````python
# Обычная функция
def square(a):                  # Функция square принимает параметр а 
  return a**2                   # И возвращает число, возведенное в квадрат

# Лямбда-функция
square_lambda = lambda a: a**2  # Функция square_lambda возведет переменную а в квадрат
                                # Можно не присваивать имя, просто определить lambda a: a**2
                                # В таком случае  итог выражения (lambda a: a**2) будер равен результату a**2

a = 2
print('square(a) = ', square(a), type(square))                      # Смотрим работу обычной функции
print('square_lambda(a) = ', square_lambda(a), type(square_lambda)) # Смотрим работу lambda функции
````

**Создание и выполнение лямбда-функций**

````python
# Способ №1
square_lambda = lambda a: a**2   # Создаем lambda функцию с именем square_lambda
print(square_lambda(a))          # Вызываем square_lambda и подаем в нее переменную а

# Способ №2
print((lambda a: a**2)(a))       # Создаем неименованную lambda функцию (lambda a: a**2) и подаем в нее переменную (a)
````

**Классические конструкции с лямбда-функциями**

Функции `map()` и `filter()` встроены в язык Python и позволяют применять для некоторого набора данных функцию `lambda` для каждого конкретного элемента в этом наборе.

Функция `filter()` возвращает `True` или `False`, если элемент, указанный в лямбда-выражении, соответствует элементам набора данных. Фильтр оставляет те элементы, которые были определены условием как `True`.

````python
# В паре с map
list(map(lambda x: x.upper(),['cat','dog', 'cow']))  # Применяем функцию верхнего регистра к списку элементов
````

Функция `map()` приняла на вход список `['cat', 'dog', 'cow']`, а также функцию `x.upper()`, которую с помощью лямбда-выражения применила ко всем элементам списка. Это более короткая запись, чем создание цикла.

В конце полученные данные конвертируем в тип данных `list`, потому что иначе функция `map` вернет тип данных `map object`, с которыми неудобно работать.

````python
# В паре с filter
list(filter(lambda x: 'o' in x, ['cat','dog', 'cow']))  # Ищем элементы в списке, соответствующие условию 
````

Функция `filter` при помощи лямбда-выражения отсеивает элементы списка, не содержащие символа `'о'`. Также полученные данные переводим в тип данных `list`.

````python
# В паре с sorted
ids = ['id1','id2','id3','id30','id22','id100']

# Лексикографическая сортировка
print(sorted(ids))                                    # методом sorted сортируем в лексическом порядке и отображаем

# Порядок целочисленных индексов
sorted_ids = sorted(ids,key = lambda x: int(x[2:]) )  # сортировка по ключам, создаваемым лямбда-выражением
print(sorted_ids)                                     # отображаем элементы, отсортированные вторым способом
````

В данном примере `lambda` принимает на вход элемент списка `ids` (например, `'id100'`). Этот элемент содержится в `х`. Мы знаем, что первые два символа каждого элемента содержат буквы `id`, которые нам не нужны. Выражение `x[2:]` возьмет все элементы из `х` кроме первых двух (`'id100'` превратится в `'100'`). Получив строку с числом, переводим ее в числовой тип данных `int`. И далее сортируем элементы списка `ids` по этим числам как ключам.

**Аргументы лямбда-функций**

Подача параметров аналогична обычной `def`-функции, которую мы уже рассмотрели ранее.

````python
print((lambda x,y,z: x + y + z)(1,2,3))           # Пример позиционной подачи параметров
print((lambda x,y,z=3: x + y + z)(1,2))           # Пример значения по умолчанию
print((lambda x,y,z=3: x + y + z)(1, y=2))        # Пример именованного параметра
print((lambda *arg: arg[0] + arg[1])(1,2))        # Пример подачи набора параметров
print((lambda **kwarg: kwarg.values())(x=1,y=2))  # Пример подачи именованных параметров
````

# Итерируемые объекты и генераторы

### Что такое итерируемые объекты и в чем их главное отличие генераторов

Очень полезны объекты, которые можно перечислить, то есть последовательно считать их значения в цикле ```for...in...```

Например, когда мы создаем список, мы можем считывать его элементы один за другим. Это называется **итерацией**:
```
mylist = [1, 2, 3]
for i in mylist :
  print(i)
```

Примеры таких объектов — списки, строки, файлы.

Очень удобно, НО!

Все значения итерируемого объекта одновременно загружены в память. И что-то подсказывает, что наша машина будет не очень хорошо себя чувствовать при попытке сделать:
```
mylist = [x*x for x in range(1000000000)]
``` 

Если же при этом стоит задача все же пройтись по списку ```mylist```, нужно использовать генераторы, которые не хранят все значения в памяти, а генерируют их на лету:
```
mygenerator = (x*x for x in range(3))
for i in mygenerator :
  print(i)
```


### Общий вид функции генератора

```
def generator_name(parameters):
  generated_values = [...]
  for one_generated_value in generated_values:
    yield one_generated_value
```


*   ```generator_name``` — имя функции генератора;
*   ```parameters``` — параметры необязательны, но если 
присутствуют, то используются как параметры обыкновенных функций;
*   ```generated_values = [...]```  — список тех значений, который необходимо обойти, генерируя очередное значение;
*   ```one_generated_value``` — очередное значение, выдаваемое генератором;
*   ```yield``` — ключевое слово, несущее смысл слова ```return```.


````python
# Пример генератора

def square_num(nums):					# Создаем функцию генератора
	for i in nums:							# В цикле берем элементы i из списка nums
		yield i**2								# yield вернет значение i**2, но не прервет цикл

for i in square_num([1,2,3]): # Поочередно генерируем значения i
  print(i)										# Отображаем
````

Особенность в том, что при проходе цикла `for` запоминается наше текущее положение и после `yield` мы продолжаем идти по списку элементов. Мы возвращаем результат вычисления на текущем этапе и не храним в памяти итог всех вычислений.

**Важно**: повторно обратиться к генератору невозможно, так как значения не запоминаются.

````python
# Пример генератора

my_generator = (x*x for x in range(3))    # Создадим генератор, умножающий число само на себя для последовательности из трех чисел
print(my_generator)                       # Смотрим, что генератор создан
for i in my_generator:                    # Берем последовательно каждый элемент
  print(i)                                # И выводим результат вычислений


# Повторное обращение к генератору

# my_generator = (x*x for x in range(3))  # Специально не включаем генератор заново, чтобы убедиться, что он ничего не хранит
print(my_generator)                       # Смотрим, что генератор все еще создан          
for i in my_generator:                    # Берем последовательно каждый элемент
  print(i)                                # И ничего не получаем. Данные не сохранены
````

### Обработка исключений

**Общая конструкция** для перехвата исключений выглядит следующим образом: 
```
try:
  try_suite
except exception_group1 as variablel:
  except_suite1
...
except ...
...
except exception_groupN as variableN:
  except_suiteN
else:
  else_suite
finally:
  finally_suite
```


*   ```try, except, else, finally``` — ключевые слова;
*   ```try_suite``` — то, что пытается выполнить интерпретатор;
*   ```exception_group1 as variablel``` — ошибка 1 и ссылка на нее (необязательно);
*   ```except_suite1``` — поведение в случае возникновения ошибки 1;
*   ```exception_groupN as variablel``` — ошибка N и ссылка на нее (необязательно);
*   ```except_suiteN``` — поведение в случае возникновения ошибки N;
*   ```else_suite``` — поведение в случае не возникновения ошибки;
*   ```finally_suite``` — поведение после окончания независимо от возникновения ошибки.



**Порядок выполнения** конструкции `try...except...finally`

Согласно приведенной выше схеме: 
интерпретатор пытается выполнить блок `try_suite`. Если все в порядке, то выполняется `else_suite` блок, а затем блок `finally_suite`. В случае возникновения ошибки интерпретатор проверит каждое предложение `except`. Если найдена соответствующая группа `exception_group`, то выполнится `except_suite`. 

Соответствующей считается группа, в которой присутствует исключение того же типа, что и возникшее исключение. Или возникшее исключение — 
подкласс одного из исключений, перечисленных в группе.

Порядок_выполнения.png

Если все прошло в штатном режиме, и ошибки не совершается, мы можем указать `else` блок, который будет выполнен в этом случае.

Блок `finally` будет выполнен в конце вне зависимости от того, была ошибка или нет.

Рассмотрим работу `try ... except` на примерах:

````python
# Функция перехвата исключения

def exc_func(input_list, return_value_index):       # Создадим функцию, принимающую некоторый список и номер индекса
  try:                                              # Пробуем выполнить
    return input_list[return_value_index]           # Вернуть из функции элемент списка под указанным индексом
  except IndexError as ie:                          # Исключение: если ошибка индексации
    print('Error:', ie)                             # Выводим на экран Error: и тип ошибки (list index out of range)
  finally:                                          # В конце выполняем следующее:
    print('A finally part of try ... except ...!')  # Выводим сообщение на экран из блока finally
  
  print('A code block after try ... except ...!')   # Выводим сообщение на экран из конца кода функции
````

Теперь запустим эту функцию без ошибки и с ошибкой:

````python
# Нормальное выполнение 

my_list = [1, 2, 3]                      # Наш список
print('Нормальное выполнение')           # Выводим сообщение на экран
a = exc_func(my_list, 0)                 # Передаем в функцию exc_func список my_list и запрашиваем элемент под индексом 0
print(a)                                 # Выводим результат на экран
````

Как мы видим, ошибки не было, и последнее, что было исполнено — блок `finally`. Функция вернула результат, и дальнейший код функции уже не исполняется.

````python
# Обработанное исключение

print('Обработанное исключение')         # Выводим сообщение на экран
a = exc_func(my_list, 3)                 # Передаем в функцию exc_func список my_list и запрашиваем элемент под индексом 3
print(a)                                 # Выводим результат на экран. Результата не будет, и получим None
````

````python
# Необработанное исключение

print('Необработанное исключение')        # Выводим сообщение на экран
a = exc_func(my_list, 'index')            # Передаем в функцию exc_func список my_list и передаем некорретный индекс
````

Поскольку строка `'index'` не может быть распознана в качестве индекса списка, наша ошибка имеет тип `TypeError`, а не `IndexError`, и потому не будет поймана нашим `except`.

Блок `finally` был выполнен вне зависимости от успешности перехвата ошибки, а дальше выполнение кода было прекращено с ошибкой.

### Иерархия классов исключений в языке Python

Рассмотрим структурную иерархию исключений:


Иерархия.png

Создадим ошибку индексации `IndexError` и попробуем ее перехватить.

````python
# Функция перехвата исключения IndexError

def exc_func(input_list, return_value_index):       # Создадим функцию, принимающую некоторый список и номер индекса
  try:                                              # Пробуем выполнить
    return input_list[return_value_index]           # Вернуть из функции элемент списка под указанным индексом
  except IndexError as ie:                          # Исключение: если ошибка индексации
    print('Error:', ie)                             # Выводим на экран Error: и тип ошибки (list index out of range)
  finally:                                          # В конце выполняем следующее:
    print('A finally part of try ... except ...!')  # Выводим сообщение на экран из блока finally
  
  print('A code block after try ... except ...!')   # Выводим сообщение на экран из конца кода функции

my_list = [1, 2, 3]                                 # Наш список
a = exc_func(my_list, 3)                            # Передаем в функцию exc_func список my_list и запрашиваем элемент под индексом 3
````

Перехватили успешно, ошибку распознали как `Error: list index out of range`.

Теперь попробуем в этой же ситуации использовать `LookupError`, которая является объединяющей для `IndexError` и `KeyError`:

````python
# Функция перехвата исключения LookupError

def exc_func(input_list, return_value_index):
  try:
    return input_list[return_value_index]
  except LookupError as le:                         # Исключение: если ошибка принадлежит категории LookupError
    print('Error:', le)
  finally:
    print('A finally part of try ... except ...!')
  
  print('A code block after try ... except ...!')

my_list = [1, 2, 3]
a = exc_func(my_list, 3)
````

Ошибку успешно перехватили. Мы можем посмотреть сообщение `Error: list index out of range`, чтобы понять, что ошибка из категории `IndexError`.

Теперь попробуем применить универсальный `Exception`:

````python
# Функция перехвата исключения Exception

def exc_func(input_list, return_value_index):
  try:
    return input_list[return_value_index]
  except Exception as ee:                           # Исключение: если ошибка принадлежит категории Exception
    print('Error:', ee)
  finally:
    print('A finally part of try ... except ...!')
  
  print('A code block after try ... except ...!')

my_list = [1, 2, 3]
a = exc_func(my_list, 3)
````

Ошибка опять успешно перехвачена. Видим, что текст ошибки `Error: list index out of range` остается неизменным вне зависимости от того, чем именно осуществлен перехват.

На случай непредусмотренной ошибки мы можем использовать блок:
```
except:
    print('Unexpected error!')
```

В данном случае `except` перехватит оставшиеся исключения. Для теста будем искать заведомо неприменимый тип ошибки — `IndexError`:



````python
# Функция перехвата неожидаемого исключения

def exc_func(input_list, return_value_index):
  try:
    return input_list[return_value_index]
  except IndexError as ie:                          # Исключение: если ошибка принадлежит категории IndexError
    print('Error:', ie)
  except:
    print('Unexpected error!')
  finally:
    print('A finally part of try ... except ...!')
  
  print('A code block after try ... except ...!')

my_list = [1, 2, 3]
a = exc_func(my_list, 'index')
````

### Возбуждение исключений

Существуют способы создания своих классов исключений. При определенных нами условиях мы можем вызывать эти исключения. Удобное средство управления потоком выполнения и тестирования программы — конструкция ```raise```.

Общий вид:
```
raise exception
raise
```


*   ```raise``` — ключевое слово;
*   ```exception``` — возбуждаемое исключение.



````python
a = 2
b = 7
if a < b:                     # Зададим условие, что если a < b, то
  raise ZeroDivisionError     # Вызываем ошибку ZeroDivisionError (или любую другую, какую укажем)
````

С помощью этого инструмента мы можем отслеживать, чтобы программа работала в строго заданных параметрах и предпринимать действия, если параметры вышли за пределы допустимых.
